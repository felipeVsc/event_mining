### KNN

```python
from duck_solution import EngineDuck

engine = EngineDuck(host='localhost',port='5432',name='example',user='postgres',password='password')
engine.registerAllVisualizeFunctions()
engine.registerAllMlFunctions()
engine.registerAllSimilarityFunctions()

engine.runQuery(""" 
WITH cteA as (SELECT feature1, feature2, feature3, target_classification FROM ml_dataset WHERE target_regression > 10 AND target_regression < 20), 
cteB AS (SELECT feature1, feature2, feature3 FROM ml_dataset WHERE target_regression < 10 OR target_regression > 20),
ml AS (SELECT CLASSIFY_KNN(3,[cteA.feature1, cteA.feature2, cteA.feature3], cteA.target_classification, [cteB.feature1, cteB.feature2, cteB.feature3]) as tree FROM cteA, cteB)
SELECT tree FROM ML;
""")
```

### Decision Tree

```python
from duck_solution import EngineDuck

engine = EngineDuck(host='localhost',port='5432',name='example',user='postgres',password='password')
engine.registerAllVisualizeFunctions()
engine.registerAllMlFunctions()
engine.registerAllSimilarityFunctions()

engine.runQuery(""" 
WITH cteA as (SELECT feature1, feature2, feature3, target_classification FROM ml_dataset WHERE target_regression > 10 AND target_regression < 20), 
cteB AS (SELECT feature1, feature2, feature3 FROM ml_dataset WHERE target_regression < 10 OR target_regression > 20),
ml AS (SELECT CLASSIFY_DECISION_TREE([cteA.feature1, cteA.feature2, cteA.feature3], cteA.target_classification, [cteB.feature1, cteB.feature2, cteB.feature3], True) as tree FROM cteA, cteB)
SELECT tree FROM ML;
""")
```

### Linear Regression
